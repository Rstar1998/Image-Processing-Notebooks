```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Reading the image file
img = cv2.imread('index.jpeg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, 'gray')
```

```python
# threshold(IMAGE_FILE, THRESHOLD_VALUE, VALUE_TO_ASSIGN, TYPE_OF_THRESHOLDING)
ret,thresh1 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh1, 'gray')
```

```python
# threshold(IMAGE_FILE, THRESHOLD_VALUE, VALUE_TO_ASSIGN, TYPE_OF_THRESHOLDING)
ret,thresh2 = cv2.threshold(gray_img,100,255,cv2.THRESH_BINARY)
plt.imshow(thresh2, 'gray')
```

```python
# Bitwise not requires grayscale image and it inverts each pixel by doing 255-PIXEL_VALUE
negative = cv2.bitwise_not(gray_img)
plt.imshow(negative, 'gray')
```

```python
neg = 255-gray_img
plt.imshow(neg, 'gray')
```

```python
def adjust_gamma(image, gamma=1.0):
# build a lookup table mapping the pixel values [0, 255] to
# their adjusted gamma values
  invGamma = 1.0 / gamma
  table = np.array([((i / 255.0) ** invGamma) * 255
  for i in np.arange(0, 256)]).astype("uint8")
# apply gamma correction using the lookup table
  return cv2.LUT(image, table)
```

```python
gamma1 = adjust_gamma(gray_img, 0.5)
gamma2 = adjust_gamma(gray_img, 1)
gamma3 = adjust_gamma(gray_img, 2)
gamma4 = adjust_gamma(gray_img, 5.5)
plt.imshow(gamma1, 'gray')
```

```python
gray_slice_without = np.zeros_like(gray_img)
# Gray level slicing Without background, chosen gray range: 75 - 125, make it 255, and others zero
for i in range(gray_img.shape[0]):
    for j in range(gray_img.shape[1]):
      if gray_img[i,j] >= 75 and gray_img[i,j] <= 125:
        gray_slice_without[i,j] = 255
plt.imshow(gray_slice_without, 'gray')
```

```python
gray_slice_with = np.copy(gray_img)
```

```python
# Gray level slicing with background, chosen gray range: 75 - 125, make it 255, and others as it is
for i in range(gray_img.shape[0]):
  for j in range(gray_img.shape[1]):
    if gray_img[i,j] >= 75 and gray_img[i,j] <= 125:
      gray_slice_with[i,j] = 255
plt.imshow(gray_slice_with, 'gray')
```
