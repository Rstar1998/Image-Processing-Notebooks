```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib import pyplot as plt
#import necessary libraries
```

```python
img = cv2.imread('index.jpeg',0) # input image
plt.imshow(img,'gray') # plot image
```

```python
hist = np.zeros(256) # make a histogram
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    hist[img[i,j]]+= 1
plt.plot(hist) # plot a histogram
```

```python
peak = hist.max() # get max (peak) value from histogram
peak_intensity = np.argmax(hist)
m = peak/peak_intensity #slope = b/a
```

```python
def perp_dist(m,p,q): # get the perpendicular distance
  return (m*p-q)/math.sqrt(m*m+1)
```

```python
max_dist = 0
threshold = 0 # calculate threshold
for i in range(peak_intensity):
  temp = perp_dist(m,i,hist[i])
  if(temp>max_dist):
    max_dist = temp
    threshold = i

threshold
```

```python
plt.axvline(x=threshold, color='r', linestyle='-') # plot threshold line
plt.plot(hist) # plot histogram
```
